Handling missing values


Here null values in all the columns are 'Missing at random', so imputing the values acordingly

```python
df['V_TAT'] = df['V_TAT'].fillna(-1)   
```

```python
df['C_TAT'] = df['C_TAT'].fillna(-1)
```

```python
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('Completed/Driver Canceled')
```

```python
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('Completed/Customer Canceled')
```

```python
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('Ride not started')
```

```python
df['Incomplete_Rides_Reason'] = np.where(df['Incomplete_Rides'] == 'No', 'Not an Incomplete Ride', df['Incomplete_Rides_Reason'])
df['Incomplete_Rides_Reason'] = np.where(df['Incomplete_Rides'] == 'Ride not started', 'Ride not started', df['Incomplete_Rides_Reason'])
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna("Not Applicable")
```

```python
df['Driver_Ratings'] = df['Driver_Ratings'].fillna(-1)
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(-1)
```

```python
100*df.isnull().sum()/df.shape[0]    # Checking the null value percentage again 
```
