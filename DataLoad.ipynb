```python
import mysql.connector        # Importing mysql.connector library for connecting python with mysql
```

```python
Connection = mysql.connector.connect(
    host = 'localhost',       # Database host
    user = 'root',            # Database username
    password = '12345',       # Database password
    allow_local_infile = True
)
```

```python
cursor = Connection.cursor()      #Creating cursor object
```

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS Project_OLA;")   # Creating Database ProjectOLA
print("MySQL database 'Project_OLA' created successfully!")
```

```python
cursor.execute("USE Project_OLA;")
cursor.execute("""                                                           
CREATE TABLE IF NOT EXISTS OLA_RidesData(
Date DATE,
Time TEXT,
Booking_ID TEXT,
Booking_Status TEXT,
Customer_ID TEXT,
Vehicle_Type TEXT,
Pickup_Location TEXT,
Drop_Location TEXT,
V_TAT DOUBLE,
C_TAT DOUBLE,
Canceled_Rides_by_Customer TEXT,
Canceled_Rides_by_Driver TEXT,
Incomplete_Rides TEXT,
Incomplete_Rides_Reason TEXT,
Booking_Value INT,
Payment_Method TEXT,
Ride_Distance INT,
Driver_Ratings DOUBLE,
Customer_Rating DOUBLE,
`Vehicle Images` TEXT
);
""")                                       # Creating Table OLA_Data

Connection.commit()
print("Table 'OLA_RidesData' Created Successfully!")
```

```python
cursor.execute("USE Project_OLA;")
cursor.execute("SET GLOBAL local_infile = ON;")
query = """
LOAD DATA LOCAL INFILE 'ola_Cleaned_Data.csv'
INTO TABLE OLA_RidesData
FIELDS TERMINATED BY ',' 
OPTIONALLY ENCLOSED BY '"'
LINES TERMINATED BY '\r\n'
IGNORE 1 LINES; 
"""

cursor.execute(query)                 # Loading Data into OLA_Data
Connection.commit()
print("Data Inserted into table 'OLA_RidesData'")
```
